 ## 1. Total Sum of Squares (TSS):
The TSS measures the total variance in the observed data. It is the sum of the squared differences between the observed values and the mean of the observed values. It reflects how much the data points deviate from the overall mean.

## 2. Residual Sum of Squares (RSS):
The RSS measures the total deviation of the predicted values from the observed values. It is the sum of the squared residuals, where a residual is the difference between an observed value and its corresponding predicted value from the model.

Mathematically:

```python
import numpy as np

# Example data
y = np.array([3, 5, 2, 8, 7])        # Observed values
y_hat = np.array([2.5, 5, 4, 7, 6.5])  # Predicted values

# Calculate TSS
y_mean = np.mean(y)  # Mean of observed values
TSS = np.sum((y - y_mean) * (y - y_mean))

# Calculate RSS
RSS = np.sum((y - y_hat) * (y - y_hat))

# Output the results
print("Total Sum of Squares (TSS):", TSS)
print("Residual Sum of Squares (RSS):", RSS)
```

### Explain
TSS: Measures how much the observed data points vary from their mean. A higher TSS indicates more variability in the data.

RSS: Measures the discrepancy between the observed data and the model's predictions. A lower RSS indicates a model that better fits the data.
